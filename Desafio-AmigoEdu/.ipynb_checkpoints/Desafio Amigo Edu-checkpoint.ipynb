```python
#imports

import pandas as pd

#grafico
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lendo arquivo

tabelaA = pd.read_csv("E:\\GIT\\PowerBI\\Desafio-AmigoEdu\\datasetCSV\\table_A.csv")
tabelaA.head()
```

```python
tabelaB = pd.read_csv("E:\\GIT\\PowerBI\\Desafio-AmigoEdu\\datasetCSV\\table_B.csv")
tabelaB.head()
```

```python
#realizando o join das duas tabelas

dfData = tabelaB.join(tabelaA.set_index('convId'), on='convId')
dfData.head()
```

```python
dfData.info()
```

```python
#quantidade de linhas

len(dfData)
```

```python
#converter createdAt em data, pois esta como objtec-string
dfData['Date'] = pd.to_datetime(dfData['createdAt'], infer_datetime_format=True)
dfData.info()
```

```python
#Apagando a coluna index gerada a partir do join

dfData.drop('index', 1, inplace=True)
dfData
```

```python
# Apagando valores NAN = valoes não identificados

dfdata = dfData.dropna()
len(dfdata)
```

```python
#Exploração de dados

#grafico
sns.pairplot(dfdata, hue='channel', height=4)
```

```python
sns.lmplot(data = dfdata, x = 'createdAt', y = 'revenue', hue = 'channel', height = 10, fit_reg= False)
sns.set_style('whitegrid')
```

```python
dfChannel = dfdata.groupby('channel').sum().sort_values(by='revenue', ascending = False)
dfChannel

#
```

```python
dfChannel = dfdata.groupby('channel').mean().sort_values(by='revenue', ascending = False)
dfChannel

# podemos observar que a conversão  de alguns canais é baixa, porém, o valor médio da venda é alto.
# implentando estratégias para melhorar a conversão, podemos aumentar a média da mesma e subir o valor médio da venda
# (com muito mais lucro)
```

```python
#multipliquei o valor da conversão por 100 para transformar em porcetagem e ter uma vizualição gráfica da minha afirmação acima
dfChannel['contribution'] = dfChannel['contribution'] * 100
dfChannel
```

```python
sns.countplot(x='channel', data=dfdata)
```

```python
#agora crie novas colunas para hora, mes e dia da semana usando a função time, apply e lambda
dfdata['Month'] = dfdata['Date'].apply(lambda time: time.month)
dfdata['Day of Week'] = dfdata['Date'].apply(lambda time: time.dayofweek)
dfdata['year'] = dfdata['Date'].apply(lambda time: time.year)
```

```python

#Agora use o .map criar um dicionario e nomear o dia da semana em str = domingo, segunda...

dmap = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}

dfdata['Day of Week'] = dfdata['Day of Week'].map(dmap)

dfdata.head()
```

```python

#Agora use o .map criar um dicionario e nomear o dia da semana em str = domingo, segunda...

dmap = {1:'jan', 2:'fev', 3:'march', 4:'april', 5:'may', 6:'jun', 7:'jul', 8:'ago', 9:'sept', 10:'ouct', 11:'nov', 12:'dez'}

dfdata['Month'] = dfdata['Month'].map(dmap)

dfdata.head()

```

```python
dfdata.drop('createdAt', 1, inplace = True)
dfdata.head()
```

```python
#Crie um grafico usando seaborn countplot em que o dia da semana estabelaça uma co-relação com a razão da ligação

sns.countplot(x='revenue', data=dfdata, hue='Day of Week')
```

```python
sns.countplot(x='revenue', data=dfdata, hue='Month')
```
